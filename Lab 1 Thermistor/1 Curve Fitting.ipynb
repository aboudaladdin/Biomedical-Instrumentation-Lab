```python
import numpy as np
from numpy.polynomial import Polynomial
import matplotlib.pyplot as plt
```

```python
# Thermistor 1
temprature = [
22,
25.38,
27.38,
30.25,
32.25,
35.19,
37.13,
40,
41.94,
44.81,
]

vo_averge = [
0.0015,
0.116,
0.208,
0.342,
0.4275,
0.5555,
0.634,
0.753,
0.8215,
0.927,
]

# Thermistor 2
# temprature = [
# 22,
# 25.31,
# 27.38,
# 30.25,
# 32.19,
# 35.13,
# 37.06,
# 39.94,
# 41.94,
# 44.81,
# ]

# vo_averge = [
# 0.004,
# 0.1425,
# 0.2255,
# 0.3395,
# 0.4025,
# 0.5025,
# 0.567,
# 0.666,
# 0.7095,
# 0.809,
# ]
```

```python
plt.plot(vo_averge, temprature, 'b', label='Ground Truth')
plt.ylabel("Temprature")
plt.xlabel("Vo")
plt.grid()
plt.legend()
plt.title("Data")
plt.show()
```

$$T = a * V_{out} + b $$

```python
# fitting 1st degree polynomial 
z = Polynomial.fit(vo_averge, temprature, 1)
coef = z.convert().coef
print(f"Coefficents : {coef}")
```

```python
plt.figure(figsize=(12,12))

plt.plot(vo_averge, temprature, 'b', label='Ground Truth')
fitting = [(v*coef[1]) + coef[0] for v in vo_averge]
plt.plot(vo_averge, fitting, 'r--', label='Fitting')
plt.ylabel("Temperature")
plt.xlabel("Vo")
plt.grid()
plt.legend()
plt.title("1D Polynomial Fitting")
plt.show()
```
